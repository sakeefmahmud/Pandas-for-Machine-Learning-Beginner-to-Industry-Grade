Detect Outliers using IQR

IQR (Interquartile Range) method is common in ML.

```python
Q1 = df['SAT_Average'].quantile(0.25)
Q3 = df['SAT_Average'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['SAT_Average'] < lower_bound) | (df['SAT_Average'] > upper_bound)]
print(outliers.shape)
```

Remove Outliers

```python
df = df[(df['SAT_Average'] >= lower_bound) & (df['SAT_Average'] <= upper_bound)]
```
